# 03 Databases / Query an HDF5 file

HDF5 can be considered a database itself. The toolbox implements `find`-methods to detect dataset or objects based on mongodb-like syntax:

```python
import h5rdmtoolbox as h5tbx
```

```python
with h5tbx.File('vortex_pair.hdf') as h5:
    h5.dump()
```

```python
with h5tbx.File('vortex_pair.hdf') as h5:
    res = h5.find({'units': {'$exists': True}})
```

```python
res
```

```python
with h5tbx.File('vortex_pair.hdf') as h5:
    res = h5.find_one({'units': 'm/s'})
```

```python
res
```

```python
with h5tbx.File('vortex_pair.hdf') as h5:
    res = h5.find({'units': {'$regex': '^m/'}})
```

```python
res
```

```python
with h5tbx.File('vortex_pair.hdf') as h5:
    res = h5.find({'$basename': 'u'})
```

```python
res
```

```python
with h5tbx.File('vortex_pair.hdf') as h5:
    res = h5.find({'standard_name': 'x_velocity'})
```
